```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

# Load the Iris dataset
iris_data = load_iris()
features = iris_data.data
labels = iris_data.target
feature_labels = iris_data.feature_names
class_names = iris_data.target_names

# Split data
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.3, random_state=42)

# Train model
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)

# Predict
predictions = classifier.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, predictions))
print(classification_report(y_test, predictions, target_names=class_names))

# Plot using petal length (index 2) and petal width (index 3)
df_iris = pd.DataFrame(features, columns=feature_labels)
plt.figure(figsize=(10, 6))
for i in range(3):
    plt.scatter(df_iris.iloc[labels == i, 2], df_iris.iloc[labels == i, 3], label=class_names[i])
plt.xlabel(feature_labels[2])  # Petal Length
plt.ylabel(feature_labels[3])  # Petal Width
plt.legend()
plt.title("Iris Dataset - Petal Features")
plt.grid()
plt.show()
```
